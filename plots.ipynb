```python
import pandas as pd
data_ = {
  "Models":[t[0] for t in direct_indirect_all],
  "Type":  [t[1] for t in direct_indirect_all],
 "Accuracy":  [t[-2] for t in direct_indirect_all],
 "Avg Time/Question(s)":  [t[-1] for t in direct_indirect_all],

}
#load data into a DataFrame object:
df_direct_indirect_all_duration = pd.DataFrame(data_)
df_direct_indirect_all_duration=df_direct_indirect_all_duration.groupby(['Models'])['Avg Time/Question(s)'].mean().reset_index()   
df_direct_indirect_all_duration.head()
```

```python
import plotly.express as px
template='ggplot2'
df = df_direct_indirect_all_duration.sort_values("Avg Time/Question(s)")
fig = px.bar(df, x="Models", y="Avg Time/Question(s)",
              barmode='group',template=template,text_auto=False,
             height=300)
fig.show()
fig.write_image("images/duration.png")
```
